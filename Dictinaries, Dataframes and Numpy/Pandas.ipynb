# DataFrames and Pandas
A **dataframe** is a data structure constructed with rows and columns, similar to a database or Excel spreadsheet. It consists of a dictionary of lists in which the list each have their own identifiers or keys, such as “last name” or “food group.”

**What is Pandas?**
Pandas is a Python library used for working with data sets.
It has functions for analyzing, cleaning, exploring, and manipulating data.

Pandas allows us to analyze big data and make conclusions based on statistical theories.
Pandas can clean messy data sets, and make them readable and relevant.
Relevant data is very important in data science.

**selecting specific rows and columns**

```python
# refer to the row index:
titanic.loc[0] #you index bu number in loc here since the index values are numbers
```

```python
# Return row 0 and 1:
titanic.loc[[0,1]]
```

```python
"""subset of both rows and columns is made in one go and just using selection brackets [] is not sufficient anymore. 
The loc/iloc operators are required in front of the selection brackets []. 
When using loc/iloc, the part before the comma is the rows you want, 
and the part after the comma is the columns you want to select."""
adult_names= titanic.loc[titanic["Age"]>21, "Name"]
print(adult_names.head())

adult_names= titanic.loc[titanic["Age"]>21, ["Name","Age"]]
print(adult_names.head())
```

```python
"""When using the column names, row labels or a condition expression, use the loc operator in front of the selection brackets []. 
For both the part before and after the comma, you can use a single label, a list of labels, a slice of labels, a conditional expression or a colon.
Using a colon specifies you want to select all rows or columns."""

# When specifically interested in certain rows and/or columns based on their position in the table, use the iloc operator 
titanic.iloc[9:25, 2:5]
```

```python
# When selecting specific rows and/or columns with loc or iloc, new values can be assigned to the selected data.
titanic.iloc[0:3, 2] = 0
titanic.head()
```

```python
# Remove all rows with NULL values:
data = pd.read_csv('data.csv')
data.dropna(inplace = True)
data
```

```python
#  Replace NULL values with the number 130:
df = pd.read_csv('data.csv')
df.fillna(130, inplace = True)
df
```

```python
df = pd.read_csv('data.csv')
x = df["Calories"].mean()
print(x)
df["Calories"].fillna(x, inplace = True)
df.iloc[17:29, 4]
```

```python
# replace any empty values with mode
df = pd.read_csv('data.csv')
x = df["Calories"].mode()[0]
df["Calories"].fillna(x, inplace = True)
df
```

```python
# Remove rows with a NULL value in the "Date" column:
df = pd.read_csv("data.csv")
df.dropna(subset=['Date'], inplace = True)
df.iloc[20:24]
```

```python
# Replacing Values
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
```

```python
# to determine if there is a duplicate
print(df.duplicated())
```

```python
# Remove all duplicates
df.drop_duplicates(inplace = True)
```
