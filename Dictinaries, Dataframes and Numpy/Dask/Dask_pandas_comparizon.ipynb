```python
file_path = "college_admissions.csv"
df_dask = dd.read_csv(file_path)
df_pandas = pd.read_csv(file_path)

# Grouping Pandas
start_time = time.time()
grouped_pandas = df_pandas.groupby('tier')['attend'].mean()
print("Pandas grouping and aggregation time:", time.time() - start_time)

# Grouping with Dask
start_time = time.time()
grouped_dask = df_dask.groupby('tier')['attend'].mean().compute()  # Must call compute
print("Dask grouping and aggregation time:", time.time() - start_time)
```

```python
file_path = "Oil and Gas .csv"
# file_path = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
# df = pd.read_csv(url,sep ='\t')
# Load the dataset with Pandas
start_time = time.time()
df_pandas = pd.read_csv(file_path)
print("Pandas load time:", time.time() - start_time)

# Load the dataset with Dask
start_time = time.time()
df_dask = dd.read_csv(file_path, assume_missing=True)
print("Dask load time:", time.time() - start_time)

```

```python

# Load dataset with Dask
dask_df = dd.read_csv('Oil and Gas .csv')

# Load dataset with Pandas for comparison
pandas_df = pd.read_csv('Oil and Gas .csv')
# Group by 'id' and calculate mean 'oil_price_2000'
start_time = time.time()
dask_grouped = dask_df.groupby('id')['oil_price_2000'].mean().compute()  # 'compute()' triggers the execution
print("\nDask grouped DataFrame (mean oil_price_2000 by id)")
print(dask_grouped)
print(f"Dask grouping time: {time.time() - start_time} seconds")

start_time = time.time()
pandas_grouped = pandas_df.groupby('id')['oil_price_2000'].mean()
print("\nPandas grouped DataFrame (mean oil_price_2000 by id)")
print(pandas_grouped)
print(f"Pandas grouping time: {time.time() - start_time} seconds")
```
